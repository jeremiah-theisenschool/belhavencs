```python
model = LinearRegression(fit_intercept=True)


r1model = model.fit(np.reshape(r1trimmedtimes, (-1, 1)),  np.reshape(r1trimmedpositions, (-1, 1))) 

r1Y = r1model.predict(np.reshape(r1trimmedtimes, (-1, 1)))   
r1Ycoef = r1model.coef_[0][0]
r1Yrsquared = sklearn.metrics.r2_score(r1trimmedpositions, r1Y) 

print(r1model.coef_)

plt.scatter(r1trimmedtimes, r1trimmedpositions, c='g', marker='s', s=4)
plt.plot(r1trimmedtimes, r1Y, 'red')
plt.xlabel('Time (s)') 
plt.ylabel('Position (m)') 
plt.legend(['Robot 1 Position', 'Line Of Best Fit']) 
plt.title('All Robot 1 Positions')
```

```python
r1lines = [] 
for i in range(30): 
    x = np.reshape(list(r1trimmed[i]['Latest: Time (s)']), (-1, 1))
    y = np.reshape(list(r1trimmed[i]['Latest: Position (m)']), (-1, 1)) 
    m = LinearRegression(fit_intercept=True).fit(x, y)  
    l = m.predict(x) 
    r1lines.append(l)

##for i in range(30): 
##    plt.plot(r1trimmedtimes[0 : 71], r1lines[i])

def makeline(c): 
    t = 0 
    y = []
    for i in range(100):  
        t = 0.05 + i*0.05
        p = c*t
        y.append(p)
    return y

r1linesnointercept = []  
r1coefs = [] 
r1rsquared = []
for i in range(30): 
    x = np.reshape(list(r1trimmed[i]['Latest: Time (s)']), (-1, 1))
    y = np.reshape(list(r1trimmed[i]['Latest: Position (m)']), (-1, 1)) 
    m = LinearRegression(fit_intercept=True).fit(x, y)  
    c = float(m.coef_[0][0]) 
    l = makeline(c)
    r1linesnointercept.append(l)  
    r1coefs.append(c)  
    r1rsquared.append(sklearn.metrics.r2_score(y, l[9 : 80])) 

print(r1rsquared)
    
```
