```python
r1lines = [] 
for i in range(30): 
    x = np.reshape(list(r1trimmed[i]['Latest: Time (s)']), (-1, 1))
    y = np.reshape(list(r1trimmed[i]['Latest: Position (m)']), (-1, 1)) 
    m = LinearRegression(fit_intercept=True).fit(x, y)  
    l = m.predict(x) 
    r1lines.append(l)

##for i in range(30): 
##    plt.plot(r1trimmedtimes[0 : 71], r1lines[i])

def makeline(c): 
    t = 0 
    y = []
    for i in range(100):  
        t = 0.05 + i*0.05
        p = c*t
        y.append(p)
    return y

r1linesnointercept = []  
r1coefs = [] 
r1rsquared = []
for i in range(30): 
    x = np.reshape(list(r1trimmed[i]['Latest: Time (s)']), (-1, 1))
    y = np.reshape(list(r1trimmed[i]['Latest: Position (m)']), (-1, 1)) 
    m = LinearRegression(fit_intercept=True).fit(x, y)  
    c = float(m.coef_) 
    l = makeline(c)
    r1linesnointercept.append(l)  
    r1coefs.append(c)  
    r1rsquared.append(l.)
    
```
