```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/tips.csv')

data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['day'], data['tip'])
plt.title('Scatter Plot')
plt.xlabel('Day of Week')
plt.ylabel('Tips')
plt.show()
```

```python
plt.scatter(data['day'], data['tip'], c=data['size'], s=data['total_bill'])
plt.title('Scatter Plot')
plt.xlabel('Day of Week')
plt.ylabel('Tips')
```

```python
plt.plot(data['tip'])
plt.plot(data['size'])
plt.title('Line Plot')
plt.xlabel('Number of response')
plt.ylabel('Tips')
```

```python
plt.bar(data['day'], data['tip'])
plt.title('Bar Chart')
plt.xlabel('Day of Week')
plt.ylabel('Tips')
```

```python
plt.hist(data['total_bill'])
```

```python
import seaborn as sns
```

```python
sns.lineplot(x='sex', y='total_bill', data=data)
plt.title('Line Plot - Title From Matplotlib')
```

```python
sns.scatterplot(x='day', y='tip', data=data)
```

```python
sns.scatterplot(x='day', y='tip', data=data, hue='sex')
```

```python
sns.lineplot(x='day', y='tip', data=data)
```

```python
sns.barplot(x='day', y='tip', data=data, hue='sex')
```

```python
sns.histplot(x='total_bill', data=data,  hue='sex', kde=True)
```

```python
import plotly.express as px
fig = px.scatter(data, x='day', y='tip', color='sex')
fig.show()
```

```python
fig = px.line(data, y='tip', color='sex')
fig.show()
```

```python
fig = px.bar(data, x='day', y='tip', color='sex')
fig.show()
```

```python
fig = px.histogram(data, x='total_bill', color='size')
fig.show()
```

```python
import plotly.graph_objects as px
plot = px.Figure(data=[px.Scatter(x=data['day'], y=data['tip'], mode='markers')])
#dropdown
plot.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=["type", "scatter"],
                    label ="Scatter Plot",
                    method = "restyle"
                )

                ,

                dict(
                    args=["type", "bar"],
                    label='Bar Chart',
                    method='restyle'

                )
            ])
        )
    ]

)
```
