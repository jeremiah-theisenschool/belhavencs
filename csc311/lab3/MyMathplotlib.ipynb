```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
pokemon = pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Pokemon.csv") 
leg = pokemon[pokemon['Legendary'] == True]
nonleg = pokemon[pokemon['Legendary'] == False] 
```

```python
plt.plot(leg['HP'], leg['Attack'], 'o') 
plt.plot(nonleg['HP'], nonleg['Attack'], 'x')
```

```python
plt.hexbin(pokemon['HP'], pokemon['Defense'])
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40) 
plt.hist(pokemon['HP'], **kwargs) 
plt.hist(pokemon['Attack'], **kwargs) 
plt.hist(pokemon['Defense'], **kwargs)
```

```python
ad = pokemon['Attack'], pokemon['Defense'], pokemon['Speed'], pokemon['HP']
sns.displot(pokemon, x='Attack', y='Defense', hue='Generation');
```

```python
sns.jointplot(pokemon, x='Attack', y='HP', hue='Type 1')
```

```python
fig = px.scatter(pokemon, x='Generation', y='Attack', color='Type 1')
fig.show()
```

```python
fig = px.histogram(pokemon, x='Type 1', color='Generation')
fig.show()
```
