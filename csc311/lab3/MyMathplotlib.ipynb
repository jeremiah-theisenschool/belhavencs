```python
import pandas as pd

data = pd.read_csv("https://github.com/TienChih/tbil-stats/raw/main/data/possum.csv")

data
```

```python
import matplotlib.pyplot as plt 
import numpy as np
plt.scatter(data['head_l'], data['skull_w'])
plt.title('Scatter Plot') 
m, b = np.polyfit(data['head_l'], data['skull_w'], 1)
plt.plot(data['head_l'], m*data['head_l'] + b)
plt.xlabel('Length of Head')
plt.ylabel('Width of Skull')
plt.show()
```

```python
plt.scatter(data['age'], data['skull_w'], c=data['site'])
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Skull Width')
plt.show()
```

```python
import seaborn as sns
sns.histplot(x='tail_l', data=data,  hue='sex', kde=True)
```
